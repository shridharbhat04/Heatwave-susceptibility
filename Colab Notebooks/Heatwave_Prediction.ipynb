```python
import pandas as pd
```

```python
df = pd.read_csv('/content/GlobalLandTemperaturesByMajorCity.csv')
df.head()
```

```python
df = df.drop(columns=['AverageTemperatureUncertainty'])
df.head()
```

```python
df['dt'] = pd.to_datetime(df['dt'], dayfirst=True, errors='coerce')
df.tail()
```

```python
df['Year'] = df['dt'].dt.year
```

```python
result = df.groupby(['Year', 'City', 'Country','Latitude','Longitude']).agg({
    'AverageTemperature': 'mean'
}).reset_index()

result
```
